```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.mixture import GaussianMixture
```

```python
data=pd.read_csv("CC GENERAL.csv")
data.head()
```

```python
data=data.drop("CUST_ID", axis=1)
```

```python
data.isnull().sum()
```

```python
data.loc[(data['MINIMUM_PAYMENTS'].isnull()==True),'MINIMUM_PAYMENTS'] = data['MINIMUM_PAYMENTS'].median()
data.loc[(data['CREDIT_LIMIT'].isnull()==True),'CREDIT_LIMIT'] = data['CREDIT_LIMIT'].median()
```

```python
plt.figure(figsize = (12, 12))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', 
            xticklabels=data.columns,
            yticklabels=data.columns)
```

```python
from sklearn.cluster import AgglomerativeClustering #Importing our clustering algorithm : Agglomerative
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(data_p)  #Applying agglomerative algorithm with 5 clusters, using euclidean distance as a metric
```

```python
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

```python
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data_p['pca1'] , data_p["pca2"] , c= agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("pca1")
ax.set_ylabel("pca2")
plt.colorbar(scatter)
```

```python
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(data_p, method="complete"))
```

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=5, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(data_p)
```

```python
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
data_p["label"]=labels
```

```python
kmeans.predict(data_p)
print(kmeans.cluster_centers_)   #Printing the coordinates of cluster centers.
```

```python
plt.scatter(data_p["pca1"][data_p.label == 0],          
            data_p["pca2"][data_p.label == 0],s=80,c='magenta',label='Careful')
plt.scatter(data_p["pca1"][data_p.label == 1],
           data_p["pca2"][data_p.label == 1],s=80,c='yellow',label='Standard')
plt.scatter(data_p["pca1"][data_p.label == 2],
           data_p["pca2"][data_p.label == 2],s=80,c='green',label='Target')
plt.scatter(data_p["pca1"][data_p.label == 3], 
           data_p["pca2"][data_p.label == 3],s=80,c='cyan',label='Careless')
plt.scatter(data_p["pca1"][data_p.label == 4], 
           data_p["pca2"][data_p.label == 4],s=80,c='burlywood',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('pca1')
plt.ylabel('pca2')
plt.legend()
plt.show()
```

```python
sum_of_squared_distances=[]
K=range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(data_p)
    sum_of_squared_distances.append(km.inertia_)
```

```python
plt.plot(K,sum_of_squared_distances)
plt.xlabel("K")
plt.ylabel("sum_of_squared_dustances")
plt.title("elbow method for optimal k")
plt.show()
```

```python
# other metric validation...
# silhouette analysis : The silhouette value is a measure of how similar an object is to its own cluster (cohesion) compared to other clusters (separation). The silhouette ranges from 1 to +1, where a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters. If most objects have a high value, then the clustering configuration is appropriate. If many points have a low or negative value, then the clustering configuration may have too many or too few clusters.
```
