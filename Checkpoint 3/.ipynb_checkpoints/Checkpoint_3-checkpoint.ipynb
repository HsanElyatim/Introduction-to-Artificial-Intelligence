```python
import pandas as pd
df=pd.read_csv("titanic-passengers.csv",delimiter=';')
df.head()
```

```python
print(df.isnull().sum())
```

```python
print(df["Cabin"].value_counts())
print(df["Age"].value_counts())
print(df["Embarked"].value_counts())
```

```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(df)
```

```python
#Correlation is a statistical term which in common usage refers to how close two variables are to having a linear relationship with each other.
#Features with high correlation are more linearly dependent and hence have almost the same effect on the dependent variable. So, when two features have high correlation, we can drop one of the two features.
#the function "plot_correlation_map" Generating the correlation heat-map wich will help us compare the correlation between features.
```

```python
l=list()
for i in range(df['Name'].value_counts().sum()):
    l.append(df["Name"][i][df["Name"][i].find(',')+2:df["Name"][i].find('.')])
df['Title']=l
df=df.drop("Name", axis=1)
```

```python
g = sns.catplot(x="Title",col="Sex",data=df,kind="count")
g.set_xticklabels(rotation=90)
```

```python
g = sns.FacetGrid(df, col="Title", height=2, col_wrap=3)
g.map(sns.histplot, "Fare")
```

```python
g = sns.catplot(x="Title",col="Survived",data=df,kind="count")
g.set_xticklabels(rotation=90)
```

```python
l=list()
for i in range(df['SibSp'].value_counts().sum()):
    l.append(df['SibSp'][i]+df['Parch'][i])
df['FamilySize']=l
df=df.drop(["Parch","SibSp"], axis=1)
```

```python
g = sns.catplot(x="FamilySize",col="Survived",data=df,kind="count")
g.set_xticklabels(rotation=90)
```
