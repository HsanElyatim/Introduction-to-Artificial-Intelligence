```python
objects=pd.get_dummies(df[['school_setting', 'school_type', 'classroom', 'teaching_method', 'student_id', 'gender', 'lunch']])
```

```python
df=pd.concat([df[['n_student','pretest', 'posttest']],objects],axis=1)
df.head()
```

```python
# Building model...(random forest regressor)

x=df.drop('posttest',axis=1)
y=df['posttest']

scaler=StandardScaler()
x=scaler.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)

from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor(criterion="mse", n_estimators=10)
regressor_model = rf_regressor.fit(x_train, y_train)
model_score = regressor_model.score(x_test, y_test)

print(f"Random Forest Regressor model score: {round(model_score, 2)}")
print(f"Random Forest Regressor model score (percentage): {round(model_score, 4) * 100} %") 
```
