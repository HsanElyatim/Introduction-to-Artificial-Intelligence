```python
import mlxtend
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
import seaborn as sns
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
import numpy as np
import math
```

```python
dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],

    ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],

    ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],

    ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],

   ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
```

```python
te=TransactionEncoder()
te_ary=te.fit(dataset).transform(dataset)    #Apply one-hot-encoding on our dataset
df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
df
```

```python
sns.barplot(data=df)
plt.xticks(rotation=90)
plt.title("items distribution")
plt.show()
```

```python
frequent_itemsets=apriori(df, min_support=0.6, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.7) # associate itemsets with confidence over 70%.
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```

```python
data=pd.read_csv("Market_Basket_Optimisation.csv")
data
```

```python
records = []
for i in range(0, len(data)):
    records.append([str(data.values[i,j]) for j in range(0, 20)])
    records[i] = [x for x in records[i] if x != 'nan']
```

```python
records
```

```python
te=TransactionEncoder()
te_ary=te.fit(records).transform(records)    #Apply one-hot-encoding on our dataset
arr_df=pd.DataFrame(te_ary, columns=te.columns_)  #Creating a new DataFrame from our Numpy array
arr_df
```

```python
sns.barplot(data=arr_df)
plt.xticks(rotation=90)
plt.title("items distribution")
plt.show()
```

```python
frequent_itemsets=apriori(arr_df, min_support=0.05, use_colnames=True) #Instead of column indices we can use column names.
frequent_itemsets
```

```python
association_rules(frequent_itemsets,metric="confidence",min_threshold=0.3) # associate itemsets with confidence over 30%.
```

```python
association_rules(frequent_itemsets,metric="lift",min_threshold=1.25)
```
